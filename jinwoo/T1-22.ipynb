주어진 데이터(basic2.csv)에서 주 단위 Sales의 합계를 구하고, 가장 큰 값을 가진 주와 작은 값을 가진 주의 차이를 구하시오(절대값)

```python
import pandas as pd
import numpy as np

df = pd.read_csv('C:\\Github\\Study\\jinwoo\\bigdata\\basic2.csv')
df
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df.info()
```

```python
df['year'] = df['Date'].dt.year
df['month'] = df['Date'].dt.month
df['day'] = df['Date'].dt.day
df['Week'] = df['Date'].dt.isocalendar().week
df
```

```python
df_2022 = df[df['Year'] == 2022]
result1 = df_2022.groupby('Week')
max_2022 = result1['Sales'].sum().max()
min_2022 = result1['Sales'].sum().min()
max_2022, min_2022
```

```python
df_2023 = df[df['Year'] == 2023]
result2 = df_2023.groupby('Week')
max_2023 = result2['Sales'].sum().max()
min_2023 = result2['Sales'].sum().min()
max_2023, min_2023
```

```python
max_ans = max(max_2022, max_2023)
min_ans = min(min_2022, min_2023)
print(max_ans, min_ans)
print(max_ans - min_ans)
```
