```python
# 시험환경 세팅 (코드 변경 X)
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split

def exam_data_load(df, target, id_name="", null_name=""):
    if id_name == "":
        df = df.reset_index().rename(columns={"index": "id"})
        id_name = 'id'
    else:
        id_name = id_name
    
    if null_name != "":
        df[df == null_name] = np.nan
    
    X_train, X_test = train_test_split(df, test_size=0.2, random_state=2021)
    
    y_train = X_train[[id_name, target]]
    X_train = X_train.drop(columns=[target])

    
    y_test = X_test[[id_name, target]]
    X_test = X_test.drop(columns=[target])
    return X_train, X_test, y_train, y_test 
    
df = pd.read_csv("C:\\Github\\Study\\Document\\train.csv")
X_train, X_test, y_train, y_test = exam_data_load(df, target='Survived', id_name='PassengerId')

X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

# X_train + X_test

```python
df = pd.read_csv("C:\\Github\\Study\\Document\\train.csv")
X_train, X_test, y_train, y_test = exam_data_load(df, target='Survived', id_name='PassengerId')

X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
df = pd.concat([X_train, X_test])
df.info()
```

```python
df['Age'] = df['Age'].fillna(df['Age'].mean()) # 평균값으로 채우기
df['Cabin'] = df['Cabin'].fillna('N') # 그냥 N으로 채우기
df['Embarked'] = df['Embarked'].fillna(df.mode().iloc[0]) # 최빈값으로 채우기

df.info()
```

```python
df = df.drop('Name', axis=1) # name column 삭제
df = df.drop("PassengerId", axis=1) # PassengerId column 삭제
```

```python
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder # 문자열을 숫자로 바꿔주는 라이브러리
le = LabelEncoder()
en_col = ['Sex', 'Ticket', 'Cabin', 'Embarked'] # 문자열 -> 숫자로 바꾸고 싶은 column 선택
df[en_col] = df[en_col].apply(le.fit_transform)
df.info()
```

```python
#float형은 int로 바꿔주기(astype사용)
df['Age'] = df['Age'].astype(int) 
#df['Fare'] = df['Fare'].astype(int)
df.info()
```

```python
X_train = df[:len(X_train)]
X_test = df[len(X_train):]

X_train.shape, X_test.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
model.fit(X_train,y_train['Survived'] )
```

```python
predictions = model.predict(X_test)
predictions
```

```python
result = pd.DataFrame({'PassengerId':X_test.index, 'Survived':predictions})
result
```

```python
model.score(X_test, y_test['Survived'])
```
