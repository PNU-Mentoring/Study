### 80%이상 결측치 컬럼, 삭제

```python
rows = len(df)
print(rows)

df.isnull().sum()/100 # f3만 결측치의 비율이 80%가 넘기때문에 컬럼 삭제 대상
```

```python
# 컬럼 삭제
df = df.drop(['f3'], axis=1)
df
```

### 80%미만 결측치 컬럼, city별 중앙값으로 대체 - map 사용

```python
# 도시확인
df['city'].unique()
```

```python
# 80퍼 미만의 결측치 가진 컬럼 : f1 -> 이 칼럼은 city별 평균값으로 대체
s=df[df['city']=='서울']['f1'].median()
b=df[df['city']=='부산']['f1'].median()
d=df[df['city']=='대구']['f1'].median()
k=df[df['city']=='경기']['f1'].median()

s, b, d, k
```

```python
# f1결측치 city별 중앙값으로 대체
df['f1'] = df['f1'].fillna(df['city'].map({'서울':s,'부산':b,'대구':d,'경기':k}))
df
```

### f1 평균값 결과 출력

```python
mean = df['f1'].mean()
print(mean)
```
