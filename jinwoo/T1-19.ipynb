주어진 데이터에서 2022년 월별 Sales 합계 중 가장 큰 금액과¶
2023년 월별 Sales 합계 중 가장 큰 금액의 차이를 절대값으로 구하시오.  
단 Events컬럼이 '1'인경우 80%의 Salse값만 반영함
(최종값은 소수점 반올림 후 정수 출력)

```python
import pandas as pd

df = pd.read_csv('C:\\Github\\Study\\jinwoo\\bigdata\\basic2.csv')
df
```

```python
df['Year'] = pd.DatetimeIndex(df['Date']).year
df['Month'] = pd.DatetimeIndex(df['Date']).month
df['Day'] = pd.DatetimeIndex(df['Date']).day
df
```

```python
df['Sales2'] = df.apply(lambda x: x['Sales']*0.8 if x['Events'] == 1 else x['Sales'], axis = 1)
df
```

```python
df_2022 = df[df['Year'] == 2022]
result1 = df_2022.groupby('Month')['Sales2'].sum().max()

print(result1)

```

```python
df_2023 = df[df['Year'] == 2023]
result2 = df_2023.groupby('Month')['Sales2'].sum().max()

print(result2)
```

```python
int(round(abs(result1 - result2)))
```
