# datetime 포함

```python
train = pd.read_csv('C:\\Github\\Study\\jinwoo\\train (1).csv')
test = pd.read_csv('C:\\Github\\Study\\jinwoo\\test.csv')
#display(X_train,X_test)
train = train.drop(['casual','registered'],axis=1)
display(train, test)
```

```python
train.info()
```

```python
Count= train.pop('count')
Count
```

```python
train['datetime'] = pd.to_datetime(train['datetime'])
test['datetime'] = pd.to_datetime(test['datetime'])

train['year'] = train['datetime'].dt.year
test['year'] = test['datetime'].dt.year

train['month'] = train['datetime'].dt.month
test['month'] = test['datetime'].dt.month

train['day'] = train['datetime'].dt.day
test['day'] = test['datetime'].dt.day

train['hour'] = train['datetime'].dt.hour
test['hour'] = test['datetime'].dt.hour

train['dayofweek'] = train['datetime'].dt.dayofweek
test['dayofweek'] = test['datetime'].dt.dayofweek

train = train.drop('datetime', axis=1)
test = test.drop('datetime', axis=1)

#display(train, test)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_val, y_train, y_val = train_test_split(train, Count, test_size=0.2, random_state=2023)
X_train.shape, X_val.shape, y_train.shape, y_val.shape
```

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor()
model.fit(X_train,y_train)
```

```python
pred = model.predict(X_val)
pred
```

```python
from sklearn.metrics import r2_score
r2_score(y_val,pred)
```

```python
pred_test = model.predict(test)
pred_test
```

```python
test = pd.read_csv('C:\\Github\\Study\\jinwoo\\test.csv')

result = pd.DataFrame({
    'datetime': test['datetime'],
    'count': pred_test
})
result

result.to_csv("submission.csv", index=False)
```
