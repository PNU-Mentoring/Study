```python
import pandas as pd

train = pd.read_csv('C:\\Github\\Study\\jinwoo\\Home Credit Default Risk\\application_train.csv')
test = pd.read_csv('C:\\Github\\Study\\jinwoo\\Home Credit Default Risk\\application_test.csv')
credit_card = pd.read_csv('C:\\Github\\Study\\jinwoo\\Home Credit Default Risk\\credit_card_balance.csv')

display(train, test)
```

```python
#train과 test를 합치기(column)
alldata = pd.concat([train, test])
alldata
```

```python
credit_card.info()
```

```python
credit_card.drop('NAME_CONTRACT_STATUS', axis=1, inplace=True)
```

```python
credit_num = credit_card.groupby('SK_ID_CURR').mean().reset_index()
credit_num 
```

```python
alldata = pd.merge(alldata, credit_num, how='left', on='SK_ID_CURR')
alldata
```

```python
#alldata의 target 칼럼은 따로 변수에 저장해두고, alldata에서는 삭제
target = alldata['TARGET']
alldata.drop('TARGET', axis=1, inplace=True)

alldata
```

```python
alldata.isnull().sum()[alldata.isnull().sum() > 0].index
```

## Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
for i in alldata.columns[alldata.dtypes == 'object']:
    alldata[i] = le.fit_transform(alldata[i])
alldata
```

```python
#금융 데이터의 경우 값이 섞이는 것보다는 독립적으로 처리하는게 나을 수 있기때문에 결측치를 중앙값이나 평균값이 아닌 -1로 채워줌
alldata = alldata.fillna(-1)
```
