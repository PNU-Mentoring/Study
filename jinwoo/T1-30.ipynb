T1-30. 12월 25일 결제 금액(price)은 12월 총 결제금액의 몇 %인가? (정수로 출력)

```python
import pandas as pd

df = pd.read_csv("C:\\Github\\Study\\jinwoo\\bigdata\\payment.csv")
df
```

```python
df['date'] = pd.to_datetime(df['date'], format='%Y%m%d')
df
```

```python
df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month
df['day'] = df['date'].dt.day
df
```

```python
total_sum = df[df['month'] == 12]['price'].sum()
print(total_sum)
```

```python
chris_sum = df[(df['month'] == 12) & (df['day'] == 25)]['price'].sum()
print(chris_sum)
```

```python
result = chris_sum / total_sum * 100
print(int(result))
```
