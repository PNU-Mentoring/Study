# T1-12. 상위 10개, 하위 10개 차이

주어진 데이터에서 상위 10개 국가의 접종률 평균과 하위 10개 국가의 접종률 평균을 구하고, 그 차이를 구해보세요    
(단, 100%가 넘는 접종률 제거, 소수 첫째자리까지 출력)

```python
import pandas as pd
df = pd.read_csv("C:\\Github\\Study\\Document\\covid-vaccination-vs-death_ratio.csv")
df
```

```python
country = df['country'].unique()
date = df['date'].unique()
```

```python
df['date'] = pd.to_datetime(df['date'])
latest_per_country = df.loc[df.groupby('country')['date'].idxmax()]
latest_per_country
```

```python
error = latest_per_country['ratio'] < 100
#error.sum()
latest_per_country = latest_per_country[error]
latest_per_country  
```

```python
latest_per_country = latest_per_country.sort_values(by='ratio')
high = latest_per_country['ratio'].tail(10).mean()
print(high)
low = latest_per_country['ratio'].head(10).mean()
print(low)
print(round(high-low,1))
```
