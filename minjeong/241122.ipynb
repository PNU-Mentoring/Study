```python
import pandas as pd
import numpy as np

train=pd.read_csv("C:\\Users\\kmo82\\Desktop\\Study\\minjeong\\application_train.csv\\application_train.csv")
test=pd.read_csv("C:\\Users\\kmo82\\Desktop\\Study\\minjeong\\application_test.csv\\application_test.csv")
display(train, test)
```

```python
credit_card=pd.read_csv("C:\\Users\\kmo82\\Desktop\\Study\\minjeong\\credit_card_balance.csv\\credit_card_balance.csv")
display(credit_card)
credit_card.info()
```

```python
alldata=pd.concat([train, test], axis=0)
```

```python
credit_card.drop('NAME_CONTRACT_STATUS', axis=1, inplace=True)
```

```python
credit_card.info()
```

```python
credit_num = credit_card.groupby('SK_ID_CURR').mean().reset_index()
```

```python
alldata = pd.merge(alldata, credit_num, how='left', on='SK_ID_CURR')
alldata
```

```python
target=alldata['TARGET']
alldata.drop('TARGET', axis=1, inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
for i in alldata.columns[alldata.dtypes==object]:
    alldata[i]=le.fit_transform(list(alldata[i]))
```
