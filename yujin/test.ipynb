```python
import pandas as pd

df = pd.read_csv("C:\\Users\\jyj42\\스터디2\\Study\\yujin\\datas\\Titanic.csv")
df
```

```python
q1 = df['Fare'].quantile(0.25)
q3 = df['Fare'].quantile(0.75)

iqr = q3 - q1

under = (q1 - (1.5*iqr))
upper = (q3+(1.5*iqr))

exc=df.loc[(df['Fare']<under) | (df['Fare']>upper)]
exc
```

```python
female_count = len(exc[exc['Gender']=='female'])
female_count
```
