```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("C:\\Users\\jyj42\\스터디2\\Study\\yujin\\datas\\basic2.csv")
df
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df
```

```python
df_2022 = df[df['Date'].dt.year == 2022]
df_2023 = df[df['Date'].dt.year == 2023]
```

```python
df_2022_info = df_2022['Sales'].groupby(df_2022['Date'].dt.isocalendar().week).sum()
df_2022_info
```

```python
df_2023_info = df_2023['Sales'].groupby(df_2023['Date'].dt.isocalendar().week).sum()
df_2023_info
```

```python
week_max = max(max(df_2022_info), max(df_2023_info))
week_min = min(min(df_2022_info), min(df_2023_info))
```

```python
print(abs(week_max - week_min))
```
