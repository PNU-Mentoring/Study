```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("C:\\Users\\jyj42\\스터디2\\Study\\yujin\\datas\\basic2.csv")
df
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df['Sales_new'] = df.loc[df['Events']==1]['Sales']*0.8
df['Sales_new'] = df['Sales_new'].fillna(df['Sales'])
df
```

```python
year_2022 = df.loc[df['Date'].dt.year == 2022]
year_2022_info = year_2022['Sales_new'].groupby(year_2022['Date'].dt.month).sum()
year_2022_info
```

```python
year_2023 = df.loc[df['Date'].dt.year == 2023]
year_2023_info = year_2023['Sales_new'].groupby(year_2023['Date'].dt.month).sum()
year_2023_info
```

```python
print("{:.0f}".format(abs(max(year_2022_info)-max(year_2023_info)),0))
```
