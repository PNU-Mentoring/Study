```python
import pandas as pd
import numpy as np

df = pd.read_csv("C:\\Users\\jyj42\\스터디2\\Study\\yujin\\datas\\covid-vaccination-vs-death_ratio.csv")
df
```

```python
df["date"]=pd.to_datetime(df["date"])
latest_country = df.loc[df.groupby("country")["date"].idxmax()]
latest_country
```

```python
over = latest_country.loc[latest_country["ratio"]>100].index
latest_country = latest_country.drop(index=over)
```

```python
latest_country = latest_country.sort_values(['ratio'])
latest_country
```

```python
max_mean = np.mean(latest_country["ratio"].tail(10))
min_mean = np.mean(latest_country["ratio"].head(10))
print(np.round(max_mean - min_mean,1))
```
