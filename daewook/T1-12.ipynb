```python
# 라이브러리 불러오기

import pandas as pd
import numpy as np
```

```python
# 데이터 불러오기

df = pd.read_csv('./input/covid-vaccination-vs-death_ratio.csv')
df
```

```python
df.dtypes
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
df.dtypes
```

```python
new_df = df.loc[df.groupby('country')['date'].idxmax()]
```

```python
new_df
```

```python
new_df.sort_values(by='ratio', inplace=True)
new_df
```

```python
new_df = new_df[new_df['ratio'] <= 100.0]
new_df
```

```python
higher = new_df['ratio'].tail(10).mean()
lower = new_df['ratio'].head(10).mean()
```

```python
print(round(higher - lower, 1))
```
