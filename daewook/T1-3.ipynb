```python
import pandas as pd
import sklearn
import numpy as np
```

```python
df = pd.read_csv('./bigdatacertificationkr/basic1.csv')
df
```

```python
df.isnull().sum(axis = 0)/df.shape[0]
```

```python
df = df.drop('f3', axis = 1)
df
```

```python
df['city'].unique()
```

```python
seoul = df[df['city'] == '서울']['f1'].median()
busan = df[df['city'] == '부산']['f1'].median()
daegu = df[df['city'] == '대구']['f1'].median()
gyeonggi = df[df['city'] == '경기']['f1'].median()

seoul, busan, daegu, gyeonggi
```

```python
df['f1'] = df['f1'].fillna(df['city'].map({'서울':seoul, '부산':busan, '대구':daegu, '경기':gyeonggi}))
```
