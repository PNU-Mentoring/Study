```python
df = pd.read_csv('./bigdatacertificationkr/Titanic.csv')
df
```

```python
Q1 = df['Fare'].quantile(.25)
Q3 = df['Fare'].quantile(.75)
```

```python
IQR = Q3 - Q1
IQR
```

```python
Q1 - 1.5 * IQR, Q3 + 1.5 * IQR
```
