```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
train = pd.read_csv("./input/bike-sharing-demand/train.csv")
test = pd.read_csv("./input/bike-sharing-demand/test.csv")

y_answer = train["count"]

display(train, test)
```

```python
train = train.drop(['casual', 'registered', 'count'], axis=1)

display(train, test)
```

```python
train2 = train.drop(['datetime'], axis=1)
test2 = test.drop(['datetime'], axis=1)
```

```python
display(train2, test2)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_val, y_train, y_val = train_test_split(train2, y_answer, test_size=0.2, random_state=42)
X_train.shape, X_val.shape, y_train.shape, y_val.shape
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import *

rf = RandomForestRegressor()
rf.fit(X_train, y_train)
pred = rf.predict(X_val)
r2_score(y_val, pred)
```

```python
display(train, test)
```

```python
train['datetime'] = pd.to_datetime(train['datetime'])
test['datetime'] = pd.to_datetime(test['datetime'])
```

```python
train['year'] = train['datetime'].dt.year
train['month'] = train['datetime'].dt.month
train['day'] = train['datetime'].dt.day
train['hour'] = train['datetime'].dt.hour
train['dayofweek'] = train['datetime'].dt.dayofweek

test['year'] = test['datetime'].dt.year
test['month'] = test['datetime'].dt.month
test['day'] = test['datetime'].dt.day
test['hour'] = test['datetime'].dt.hour
test['dayofweek'] = test['datetime'].dt.dayofweek

display(train, test)
```

```python
train3 = train.drop(['datetime'], axis=1)
test3 = test.drop(['datetime'], axis=1)
```

```python
X_train, X_val, y_train, y_val = train_test_split(train3, y_answer, test_size=0.2, random_state=42)
X_train.shape, X_val.shape, y_train.shape, y_val.shape
```

```python
rf = RandomForestRegressor()
rf.fit(X_train, y_train)
pred = rf.predict(X_val)
r2_score(y_val, pred)
```
