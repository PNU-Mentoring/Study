```python
import pandas as pd
import numpy as np
```

```python
train = pd.read_csv("./house-prices-advanced-regression-techniques/train.csv")
test = pd.read_csv("./house-prices-advanced-regression-techniques/test.csv")

display(train, test)
```

```python
# int64 또는 float64 타입 열 선택
train_numeric_columns = train.select_dtypes(include=['int64', 'float64']).columns
train_numeric_columns = train_numeric_columns.drop('SalePrice', errors='ignore')
test_numeric_columns = test.select_dtypes(include=['int64', 'float64']).columns
```

```python
null_columns = train.isnull().sum()[train.isnull().sum() > 0].index

# 해당 열들의 데이터 타입 출력
print(train[null_columns].dtypes)
```

```python
null_columns_test = test.isnull().sum()[test.isnull().sum() > 0].index

# 해당 열들의 데이터 타입 출력
print(test[null_columns_test].dtypes)
```

```python
# 결측값이 있는 열 중 object 타입 열만 필터링
object_null_columns = train[null_columns].select_dtypes(include='object').columns

# 선택된 열에 대해 결측값을 "null"로 채우기
train[object_null_columns] = train[object_null_columns].fillna("null")
```

```python
train.isnull().sum()[train.isnull().sum() > 0]
```

```python
# 결측값이 있는 열 중 object 타입 열만 필터링
object_null_columns_test = test[null_columns_test].select_dtypes(include='object').columns

# 선택된 열에 대해 결측값을 "null"로 채우기
test[object_null_columns_test] = test[object_null_columns_test].fillna("null")
```

```python
test.isnull().sum()[test.isnull().sum() > 0]
```

```python
# float64 타입 열만 선택하여 결측값을 중앙값으로 채우기
float_columns = train.select_dtypes(include='float64').columns
train[float_columns] = train[float_columns].fillna(train[float_columns].median())
```

```python
train.isnull().sum()[train.isnull().sum() > 0]
```

```python
float_columns_test = test.select_dtypes(include='float64').columns
test[float_columns_test] = test[float_columns_test].fillna(test[float_columns_test].median())
```

```python
test.isnull().sum()[test.isnull().sum() > 0]
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
train_en_col = train.select_dtypes(include='object').columns
train[train_en_col] = train[train_en_col].apply(le.fit_transform)

test_en_col = test.select_dtypes(include='object').columns
test[test_en_col] = test[test_en_col].apply(le.fit_transform)
```

```python
import matplotlib.pyplot as plt

# 열을 10개씩 묶어 그리기
columns_per_plot = 10
for i in range(0, len(train.columns), columns_per_plot):
    subset = train.columns[i:i + columns_per_plot]
    
    # 각 subset에 대해 박스플롯 그리기
    train[subset].plot(kind='box', subplots=True, layout=(2, 5), figsize=(15, 10), sharex=False, sharey=False)
    plt.tight_layout()
    plt.show()
```

```python
# IQR 기반 이상치 수정
for col in train_numeric_columns:
    Q1 = train[col].quantile(0.25)  # 25번째 백분위수
    Q3 = train[col].quantile(0.75)  # 75번째 백분위수
    IQR = Q3 - Q1                  # IQR 계산

    lower_bound = Q1 - 1.5 * IQR   # 하한선
    upper_bound = Q3 + 1.5 * IQR   # 상한선

    # 이상치를 하한선 또는 상한선으로 수정
    train[col] = train[col].apply(
        lambda x: lower_bound if x < lower_bound else (upper_bound if x > upper_bound else x)
    )
```

```python
# 열을 10개씩 묶어 그리기
columns_per_plot = 10
for i in range(0, len(train.columns), columns_per_plot):
    subset = train.columns[i:i + columns_per_plot]
    
    # 각 subset에 대해 박스플롯 그리기
    train[subset].plot(kind='box', subplots=True, layout=(2, 5), figsize=(15, 10), sharex=False, sharey=False)
    plt.tight_layout()
    plt.show()
```

```python
train.drop('Id', axis=1, inplace=True)
test.drop('Id', axis=1, inplace=True)
```

```python
SalePrice = train['SalePrice']
train.drop('SalePrice', axis=1, inplace=True)
```

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(n_jobs=-1)
rf.fit(train, np.log(SalePrice))
```

```python
result = rf.predict(test)
```

```python
submission = pd.read_csv("./house-prices-advanced-regression-techniques/sample_submission.csv")
submission['SalePrice'] = np.exp(result)
```

```python
# df를 CSV 파일로 저장
submission.to_csv('sub.csv', index=False)
```

```python
# 건물 연식

train['old'] = train['YrSold'] - train['YearBuilt']
test['old'] = test['YrSold'] - test['YearBuilt']
```

```python
# 건물 면적

train['totalArea'] = train['TotalBsmtSF'] + train['GrLivArea']
test['totalArea'] = test['TotalBsmtSF'] + test['GrLivArea']
```

```python
# 차고 면적 추가한 총 면적

train['groundArea'] = train['totalArea'] + train['GarageArea']
test['groundArea'] = test['totalArea'] + test['GarageArea']
```

```python
# 전체 품질 평가

train['totalQual'] = train['OverallQual'] + train['OverallCond'] * 0.65
test['totalQual'] = test['OverallQual'] + test['OverallCond'] * 0.65
```

```python
train['diffQual'] = train['totalQual'] - train['OverallCond'] * 0.65
test['diffQual'] = test['totalQual'] - test['OverallCond'] * 0.65
```

```python
rf = RandomForestRegressor(n_jobs=-1)
rf.fit(train, np.log(SalePrice))
result = rf.predict(test)

new_submission = pd.read_csv("./house-prices-advanced-regression-techniques/sample_submission.csv")
new_submission['SalePrice'] = np.exp(result)

# df를 CSV 파일로 저장
new_submission.to_csv('sub1.csv', index=False)
```
