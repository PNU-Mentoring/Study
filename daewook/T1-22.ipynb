```python
# 라이브러리 불러오기

import pandas as pd
import numpy as np
```

```python
# 데이터 불러오기

df = pd.read_csv('./bigdatacertificationkr/basic2.csv')
df
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df.info()
```

```python
df['weekofyear'] = df['Date'].dt.isocalendar().week
df['year'] = df['Date'].dt.year
df
```

```python
# resample 사용
```

```python
df.groupby(['year', 'weekofyear'])['Sales'].sum()
```

```python
max_week = df.groupby(['year', 'weekofyear'])['Sales'].sum().max()
min_week = df.groupby(['year', 'weekofyear'])['Sales'].sum().min()

print(abs(max_week - min_week))
```
