```python
# 라이브러리 불러오기

import pandas as pd
import numpy as np
```

```python
# 데이터 불러오기

df = pd.read_csv('./bigdatacertificationkr/basic2.csv')
df
```

```python
df['Sales2'] = df.loc[df['Events'] == 1]['Sales'] * 0.8
```

```python
df
```

```python
df['Sales2'].fillna(df['Sales'], inplace=True)
```

```python
df
```

```python
df['Date'] = pd.to_datetime(df['Date'])

df['year'] = df['Date'].dt.year
df['month'] = df['Date'].dt.month

df
```

```python
max2022 = df[df['year'] == 2022].groupby('month')['Sales2'].sum().max()
```

```python
max2023 = df[df['year'] == 2023].groupby('month')['Sales2'].sum().max()
```

```python
print(round(abs(max2022 - max2023)))
```
