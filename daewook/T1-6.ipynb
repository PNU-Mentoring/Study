T1-6. 결측치 제거 및 그룹 합계 Expected Questions

```python
# 라이브러리 불러오기

import pandas as pd
import sklearn
import numpy as np
```

```python
# 데이터 불러오기

df = pd.read_csv('./bigdatacertificationkr/basic1.csv')
df
```

```python
# f1컬럼 결측치 제거

dropf1_df = df.dropna(subset=['f1'])
dropf1_df
```

```python
# 그룹 합계 계산

sum_df = dropf1_df.groupby(['city', 'f2']).sum()
sum_df.loc[:, 'f1']
```

```python
# 조건에 맞는 값 출력

print(sum_df.iloc[0]['f1'])
```
