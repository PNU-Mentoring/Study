```python
import pandas as pd
import numpy as np
```

```python
train = pd.read_csv('./input/home-credit-default-risk/application_train.csv')
test = pd.read_csv('./input/home-credit-default-risk/application_test.csv')

display(train, test)
```

```python
credit_card_balance = pd.read_csv('./input/home-credit-default-risk/credit_card_balance.csv')

display(credit_card_balance)
```

```python
data = pd.concat([train, test])

display(data)
```

```python
credit_card_balance.info()
```

```python
credit_card_balance.drop(['NAME_CONTRACT_STATUS'], axis=1, inplace=True)
```

```python
credit_num = credit_card_balance.groupby('SK_ID_CURR').mean().reset_index()
credit_num
```

```python
data = pd.merge(data, credit_num, on='SK_ID_CURR', how='left')
data
```

```python
target = data['TARGET']
data.drop(['TARGET'], axis=1, inplace=True)

display(data, target)
```

```python
data.isnull().sum()[data.isnull().sum() > 0].index
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
en_col = data.select_dtypes(include='object').columns
data[en_col] = data[en_col].apply(le.fit_transform)
```

```python
data
```

```python
data.fillna(-1, inplace=True)

data.isnull().sum()[data.isnull().sum() > 0].index
```

```python
new_train = data[:len(train)]
new_train['TARGET'] = target
new_test = data[len(train):]

display(new_train, new_test)
```

```python
from catboost import CatBoostClassifier

cbc = CatBoostClassifier()
cbc.fit(new_train.drop(['TARGET'], axis=1), new_train['TARGET'])
result = cbc.predict_proba(new_test)[:,1]
sub = pd.read_csv('./input/home-credit-default-risk/sample_submission.csv')

sub.head()
```

```python
new_train['TARGET']
```

```python
sub['TARGET'] = result
sub.to_csv('home_credit.csv', index=False)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range = (0,1))

x_train = new_train.drop(['TARGET'], axis=1)

scaler.fit(x_train)
x_train = scaler.transform(x_train)
x_test = scaler.transform(new_test)

cbc.fit(x_train, new_train['TARGET'])
result = cbc.predict_proba(x_test)[:,1]

sub['TARGET'] = result
sub.to_csv('home_credit2.csv', index=False)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

x_train = new_train.drop(['TARGET'], axis=1)

scaler.fit(x_train)
x_train = scaler.transform(x_train)
x_test = scaler.transform(new_test)

cbc.fit(x_train, new_train['TARGET'])
result = cbc.predict_proba(x_test)[:,1]

sub['TARGET'] = result
sub.to_csv('home_credit3.csv', index=False)
```
