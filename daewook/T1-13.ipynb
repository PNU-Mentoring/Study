```python
# 라이브러리 불러오기

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# 데이터 불러오기

df = pd.read_csv('./input/winequality-red.csv')
df.sort_values(by='quality')
```

```python
df.corr()
```

```python
sns.heatmap(data=df.corr(), annot=True, annot_kws={"size":5})
```

```python
max_corr = abs(df.corr()['quality'][:-1]).max()
min_corr = abs(df.corr()['quality'][:-1]).min()

print(max_corr, min_corr)
```

```python
print(round(max_corr + min_corr, 2))
```
