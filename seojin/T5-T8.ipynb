### T1-5 조건에 맞는 데이터 표준편차 구하기

```python
# 라이브러리 및 데이터 불러오기
import pandas as pd
import numpy as np

data = pd.read_csv('C:\\Users\\lsj02\\OneDrive\\바탕 화면\\PNU\\멘토링\\Study\\seojin\\bigdatacertificationkr\\basic1.csv')
data
```

```python
# 조건에 맞는 데이터 (ENFJ, INFP)
enfj = data[data['f4'] == 'ENFJ']
infp = data[data['f4'] == 'INFP']
enfj, infp
```

```python
# 조건에 맞는 f1의 표준편차 (ENFJ, INFP)
enfj = enfj['f1'].std()
infp = infp['f1'].std()
enfj, infp
```

```python
# 두 표준편차 차이 절대값 출력
abs(enfj - infp)
```

### T1-8. 누적합 그리고 보간 (결측치 처리)

```python
import pandas as pd
import numpy as np

data = pd.read_csv('C:\\Users\\lsj02\\OneDrive\\바탕 화면\\PNU\\멘토링\\Study\\seojin\\bigdatacertificationkr\\basic1.csv')
data[data['f2'] == 1]['f1'].cumsum().fillna(method = 'bfill').mean()
```

```python
# 라이브러리 및 데이터 불러오기
import pandas as pd
import numpy as np

data = pd.read_csv('C:\\Users\\lsj02\\OneDrive\\바탕 화면\\PNU\\멘토링\\Study\\seojin\\bigdatacertificationkr\\basic1.csv')
data
```

```python
# 조건에 따른 누적합
data2 = data[data['f2'] == 1]['f1'].cumsum()
data2
```

```python
data2 = data2.fillna(method = 'bfill')
data2
```

```python
# 평균 출력
data2.mean()
```
