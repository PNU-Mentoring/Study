```python
import pandas as pd

train = pd.read_csv("C:\\Users\\lsj02\\OneDrive\\바탕 화면\\PNU\\멘토링\\Study\\seojin\\bigdatacertificationkr\\bikesharing\\train.csv")
test = pd.read_csv("C:\\Users\\lsj02\\OneDrive\\바탕 화면\\PNU\\멘토링\\Study\\seojin\\bigdatacertificationkr\\bikesharing\\test.csv")

target = train.pop('count')
train = train.drop(['datetime','casual', 'registered'], axis=1)
test = test.drop(['datetime'], axis=1)
display(target, train.head(), test.head())
```

```python
from sklearn.model_selection import train_test_split

X_train, X_val, y_train, y_val = train_test_split(train, target, test_size=0.2, random_state=2023)
X_train.shape, X_val.shape, y_train.shape, y_val.shape
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score

rf = RandomForestRegressor()

rf.fit(X_train, y_train)
pred = rf.predict(X_val)
r2_score(y_val, pred)
```
