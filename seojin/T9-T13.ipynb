### T1-11. min-max 기준 상하위 5% 값

### T1-12. 상위 10개, 하위 10개 차이

```python
import pandas as pd
import numpy as np

data = pd.read_csv("C:\\Users\\lsj02\\OneDrive\\바탕 화면\\PNU\\멘토링\\Study\\Document\\covid-vaccination-vs-death_ratio.csv")
data = data.groupby('country').max().sort_values('ratio')
data = data[data['ratio'] <= 100]

round(data.tail(10)['ratio'].mean() - data.head(10)['ratio'].mean(), 1)
```

### T1-13. 상관관계 구하기

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv('C:\\Users\\lsj02\\OneDrive\\바탕 화면\\PNU\\멘토링\\Study\\Document\\winequality-red.csv')
data_corr = abs(data.corr()['quality'][:-1])
round(data_corr.max() + data_corr.min(), 2)

# 히트맵 그리기
ax = sns.heatmap(data.corr(), cmap='Blues', annot=True, annot_kws={'size': 7})
```
