### T1-4. 왜도와 첨도 구하기

```python
# 라이브러리 및 데이터 불러오기
import pandas as pd
import numpy as np

data = pd.read_csv('C:\\Users\\lsj02\\OneDrive\\바탕 화면\\PNU\\멘토링\\Study\\seojin\\house-prices-advanced-regression-techniques\\train.csv')
data
```

```python
# 'SalePrice' 컬럼 왜도와 첨도 계산
s = data['SalePrice'].skew()
k = data['SalePrice'].kurt()
s, k
```

```python
# 'SalePrice' 컬럼 로그 변환
data['SalePrice'] = np.log1p(data['SalePrice'])
data['SalePrice']
```

```python
# 'SalePrice' 컬럼 왜도와 첨도 계산
log_s = data['SalePrice'].skew()
log_k = data['SalePrice'].kurt()
log_s, log_k
```

```python
# 모두 더한 다음 출력
round(s + log_s + k + log_k, 2)
```
