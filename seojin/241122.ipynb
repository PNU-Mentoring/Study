```python
import pandas as pd
import numpy as np

train = pd.read_csv('C:\\Users\\lsj02\\OneDrive\\바탕 화면\\PNU\\멘토링\\Study\\seojin\\home-credit-default-risk\\application_train.csv')
test = pd.read_csv('C:\\Users\\lsj02\\OneDrive\\바탕 화면\\PNU\\멘토링\\Study\\seojin\\home-credit-default-risk\\application_test.csv')
credit_card_balance = pd.read_csv('C:\\Users\\lsj02\\OneDrive\\바탕 화면\\PNU\\멘토링\\Study\\seojin\\home-credit-default-risk\\credit_card_balance.csv')

display(train, test, credit_card_balance)
```

```python
credit_card_balance.drop('NAME_CONTRACT_STATUS', axis=1, inplace=True)
credit_num = credit_card_balance.groupby('SK_ID_CURR').mean().reset_index()
credit_num
```

```python
alldata = pd.concat([train, test])
alldata = pd.merge(alldata, credit_num, on='SK_ID_CURR', how='left')
alldata
```

```python
# target 분리
target = alldata['TARGET']
alldata.drop(['TARGET'], axis=1, inplace=True)

alldata
```

```python
# null값 확인
null_index = alldata.isnull().sum()[alldata.isnull().sum() > 0].index
null_index
```

```python
# object -> numeric
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
for i in alldata.columns[alldata.dtypes == 'object']:
    alldata[i] = le.fit_transform(list(alldata[i]))
alldata

# 결측치 채우기
alldata.fillna(-1) # 신용데이터 - 독립적으로 동작하도록 -1로 채움
```

```python
len_train = len(train)
train = alldata[:len_train]
test = alldata[len_train:]

train['TARGET'] = target[:len_train]
```

```python
from catboost import CatBoostClassifier

target = train['TARGET']
train = train.drop('TARGET', axis=1)

cbc = CatBoostClassifier()
cbc.fit(train, target)
# 안 갚을 확률, 갚을 확률
result = cbc.predict_proba(test)[:,1]

submission = pd.read_csv('C:\\Users\\lsj02\\OneDrive\\바탕 화면\\PNU\\멘토링\\Study\\seojin\\home-credit-default-risk\\sample_submission.csv')
submission.head()
```

```python
result  # 0: 확률, 1: 확률

submission['TARGET'] = result
submission.to_csv('home-credit-default-risk\\submission.csv', index=False)
```

```python
# train, test scaling
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range = (0,1))

scaler.fit(train)
train = scaler.transform(train)
test = scaler.transform(test)
```

```python
from catboost import CatBoostClassifier

cbc = CatBoostClassifier()
cbc.fit(train, target)
# 안 갚을 확률, 갚을 확률
result = cbc.predict_proba(test)[:,1]

submission = pd.read_csv('C:\\Users\\lsj02\\OneDrive\\바탕 화면\\PNU\\멘토링\\Study\\seojin\\home-credit-default-risk\\sample_submission.csv')

submission['TARGET'] = result
submission.to_csv('home-credit-default-risk\\scaled_submission.csv', index=False)
```
